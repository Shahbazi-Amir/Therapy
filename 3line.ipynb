شنبه ۲۲ آذر ۱۴۰۴
کد:
سوال ۸ و سوال ۹ حل شد.
مانع اینکه کدکامل بیاید یادگیری معنا ندارد با هوش مصنوعی هماهنگ شد که جواب ها داده نشود وبا مسیر فکری برویم
برای فردا سوال ۱۰ ان شاالله


یادگیری ماشین وفری لنسرینگ:
خرید سرور استفتا درباره فریلنسری

فردا: کلا بررسی و ادامه با یک سرور و اکانت جدید یا فقط سرچ

مانع های ادیتوری و گیت بود برطرف شد
فردا ان شاالله بتوانم روزی ده تا تست
