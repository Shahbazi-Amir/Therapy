در قسمت ارشد ثبت نام ارشد ۱۴۰۵






مانع کدنویسی هم خود هوش مصنوعی هم عدم کدنویسی خودم چون زمانبر است.

چهارشنبه ۱۹ آذر ۱۴۰۴

کد:
سوال ۷ حل شد درباره برگشت اینت در پوشته ۳۲ بیتی
مانعی وجود نداشت 
فردا سوال هشت

یادگیری ماشین :
برای راگ به این نتیجه عرسیدم سیستم محلی لپ تاپ ۱۰۰۰ دلاری میخواهد اگر امنیت مهم است.تازه ۵۰ ساعت هم وقت گذاشتم والسلام
این مانع و نتیجه بود
اما برای فردا
اپ ورک فول با همه چی بررسی همه چی و افتادن در آن
هم زمانش شده
هم پروژه ها دیده میشود
هم مالی
هم بررسی واقعی
هم سرور میشود استفاده چند کاره کرد

کد:
سوال ۸ و سوال ۹ حل شد.
